```python
df1 = pd.read_csv('/home/hwi/study/code_test/study/01_AI_Bigdata_Programming/jeju_eda/df_card.csv')
df1.info()
df1.head()
```

- 많이 간 식당 - 매출수 - (월별 상위 3개 순위)

```python
df1['상호명(제주도민매출수)'] = df1['상호명']
grouped = df1.groupby(['분석년월'])
func = lambda g: g.sort_values(by='제주도민매출수', ascending=False)[:3].reset_index(drop=True)#[:1]
res1 = grouped.apply(func)[['상호명(제주도민매출수)']]

df1['상호명(외지인매출수)'] = df1['상호명']
grouped = df1.groupby(['분석년월'])
func = lambda g: g.sort_values(by='외지인매출수', ascending=False)[:3].reset_index(drop=True)#[:1]
res2 = grouped.apply(func)[['상호명(외지인매출수)']]

df1['상호명(전체매출수)'] = df1['상호명']
grouped = df1.groupby(['분석년월'])
func = lambda g: g.sort_values(by='전체매출수', ascending=False)[:3].reset_index(drop=True)#[:1]
res3 = grouped.apply(func)[['상호명(전체매출수)']]

result = pd.concat([res1, res2, res3], axis = 1)
result.fillna("")
```
