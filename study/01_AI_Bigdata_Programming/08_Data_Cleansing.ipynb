### <span style="color:#004f9f"> 2.5. 문자열 데이터 다루기 </span>

**[여러 개의 정보가 한 열에 담긴 데이터의 처리]**

```python
df['시군구'] = df['시군구'].str.replace('서울특별시 ','').str.strip()
df.head()
```

```python
df['시군구'].str.split(' ')
```

```python
df['시군구'].str.split(' ', expand=True)
```

```python
df.insert(2, '시군구_대범주', df['시군구'].str.split().str[0])
df.head()
```

```python
df = df.rename(columns = {'시군구':'시군구_소범주'})
df.head()
```

**[잘못 입력된 문자열 데이터의 처리]**

```python
df['가해운전자연령'].unique()
```

```python
df['가해운전자연령'] = df['가해운전자연령'].str.replace('세 이상','')
df['가해운전자연령'] = df['가해운전자연령'].str.replace('세','')
df['가해운전자연령'].unique()
```

```python
try:
    df['가해운전자연령'] = df['가해운전자연령'].astype('int')
except:
    df['가해운전자연령'] = df['가해운전자연령'].astype('float')
```
