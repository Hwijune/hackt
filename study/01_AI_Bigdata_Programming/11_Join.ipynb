### <span style="color:#004f9f"> 2.2.  .join() </span>

```python
my_dict = {}
for i in df.keys():
    col_name = i+'여부'
    df[i][col_name] = 1
    my_dict[i] = df[i][['사고번호',col_name]].set_index('사고번호')
    
print(my_dict.keys())
locals().update(my_dict)
```

```python
my_join = 대형사고
for i in my_dict.keys():
    if i=='대형사고': continue
    my_join = my_join.join(my_dict[i], how='outer')
my_join = my_join.fillna(0)
my_join = my_join.astype('int')
my_join.info()
```

### <span style="color:#004f9f"> 2.3.  .merge() </span>

```python
for i in my_dict.keys():
    my_dict[i] = my_dict[i].reset_index()
my_dict['대형사고']
```

```python
my_dict['대형사고'].merge(my_dict['음주사고'], on='사고번호', how='outer')
```

```python
my_merge = my_dict['대형사고']
for i in my_dict.keys():
    if i=='대형사고': continue
    my_merge = my_merge.merge(my_dict[i], on='사고번호', how='outer')
my_merge = my_merge.fillna(0)
my_merge = my_merge.astype(int, errors='ignore')
my_merge.info()
```
