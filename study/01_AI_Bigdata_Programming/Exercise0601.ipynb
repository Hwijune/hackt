```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/01_seoul_accident_raw.csv', encoding='euc-kr')
df.head()
```

```python
df.info()
```

- 새로운 변수 생성
    - 사고년도
    - 사고월
    - 사고일
    - 사고시각
    - 사고요일(숫자로 표현)

```python
print(df['사고일시'])
```

```python
#데이터 타입을 object에서 datetime으로 변환
df['사고일시'] = pd.to_datetime(df['사고일시'], format='%Y년 %m월 %d일 %H시')
print(df['사고일시'])
```

```python
#새로운 변수 생성
#사고년도
df['사고년도'] = df['사고일시'].dt.year
#사고월
df['사고월'] = df['사고일시'].dt.month
#사고일
df['사고일'] = df['사고일시'].dt.day
#사고시각
df['사고시각'] = df['사고일시'].dt.time
#사고요일
df['사고요일'] = df['사고일시'].dt.weekday

df.info()
```

```python
df.head()
```

```python
#"사고일시" 컬럼 제거
col = ['사고일시']
df = df.drop(columns=col)
df.head()
```
