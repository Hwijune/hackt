```python
from flask import Flask, jsonify, request
from flask_ngrok import run_with_ngrok
app = Flask(__name__)
run_with_ngrok(app)
@app.route('/weather')
def get_weather():
  location = request.args.get('location')
  return jsonify({"날씨": "맑음", "기온": 5, "장소": location})
app.run()
```
