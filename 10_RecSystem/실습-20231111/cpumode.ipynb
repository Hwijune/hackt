```python
res_edu_dict = resume.set_index('resume_seq')['degree'].to_dict()
req_edu_dict = req.set_index('recruitment_seq')['education'].to_dict()
```

```python
res_edu_dict = dict.fromkeys(res_edu_dict, 10)
```

```python
apply_train['req_edu']=apply_train['recruitment_seq'].map(req_edu_dict)
apply_train
```

```python
for idx, row in apply_train.iterrows():
  res_edu_dict[row['resume_seq']]=min(res_edu_dict[row['resume_seq']],row['req_edu'])
```

```python
failcount=0
def isThisAcceptable(uid,itemid):
  global failcount
  if res_edu_dict[uid] <= req_edu_dict[itemid]:
    return True
  else:
    failcount+=1
    return False

  # res_edu = res_edu_dict[uid]
  # req_edu = req_edu_dict[itemid]
  # # print(res_edu)
  # # print(req_edu)
  # if res_edu >=(req_edu-1):
  #   return True
  # else:
  #   # print("FALSE")
  #   failecount+=1
  #   return False
```

```python
users, items = [], []

MYTOPK=5

for name, group in mydf.groupby('userID'):
  mydic=dict()
  found = False
  for index, row in group.iterrows():
    for j in row[1:].values:
      if  isThisAcceptable(name,j):
        if j in mydic.keys():
          mydic[j]+=1
        else:
          mydic[j]=1
      sorted_dict = dict(sorted(mydic.items(), key=lambda item: item[1], reverse=True))
      if (len(sorted_dict)>MYTOPK) and (list(sorted_dict.values())[MYTOPK-1]==target_colnum):
        mylist = list(sorted_dict.keys())[:MYTOPK]

        found = True
        break
    if found == True:
      break
  user=[name] * MYTOPK
  users.extend(user)
  items.extend(mylist)


all_predictions = pd.DataFrame(
        data={'resume_seq': users, 'recruitment_seq': items}
    )
all_predictions.to_csv(f'{FILE_PREFIX}/temp_result/go_voted.csv', index=False,header=True)
print(failcount)
```
