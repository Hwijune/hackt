```python
#!pip install pandas_datareader
# https://pandas-datareader.readthedocs.io/en/latest/remote_data.html
import pandas_datareader.data as web
# web.DataReader('종목번호','naver', start='시작일', end='종료일')
df = web.DataReader('005930', 'naver',
                    start='2023-04-01',
                    end = '2023-06-12')
df = df[['Close','Volume']]
df = df.astype('int')
df = df.reindex(pd.date_range('2023-04-01','2023-06-12'))
df = df.interpolate()
df = df.dropna()
df = df.astype('int')

news = pd.read_csv('/content/drive/Othercomputers/내 노트북/study/08_DS/news_vector.csv')
news['일자'] = pd.to_datetime(news['일자'], format='%Y%m%d')
news['일자'] = news['일자'] + pd.Timedelta(days=1)
news = news.set_index('일자')
news
```

## 2. Date Correction for Prediction

```python
df['t1'] = df['Close'].shift(1)
df['Close'] = df['Close'] - df['Close'].shift(1)
df['Volume'] = df['Volume'].shift(1)
df['mw5'] = df['Close'].rolling(5).mean().shift(1)
df['emw5'] = df['Close'].ewm(5).mean().shift(1)
df = df.dropna()
df
```

## 3. Sample Modeling

```python
train_idx = df.iloc[:-5].index
test_idx = df.iloc[-5:].index
```

```python
full_df = df
```

```python
x_train = full_df.loc[train_idx,'Volume':]
y_train = full_df.loc[train_idx, 'Close']
x_test = full_df.loc[test_idx,'Volume':]
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
pred = pd.DataFrame({'일자': x_test.index, 'pred':rf.predict(x_test)})
pred = pred.groupby('일자')['pred'].mean()
```

```python
pred.sub(df.loc[test_idx,'Close'], axis=0).abs().mean()
```
