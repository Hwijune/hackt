```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid', font_scale=1.5) #_theme 이거는 구버전
sns.set_palette('Set2', n_colors=10)
plt.rc('font', family='NanumGothic')
plt.rc('axes', unicode_minus=False)
```

```python
df = pd.read_csv('data/03_seoul_accident_full.csv')
df.head()
```

## <span style="color:#004f9f"> 1. coutplot: 범주형 변수의 빈도 시각화</span>

```python
plt.figure(figsize=(15,10))
sns.countplot(x='사고내용', data=df);
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='뺑소니사고여부', data=df, hue='가해운전자남성여부');
```

## <span style="color:#004f9f"> 2. Barplot: 수치형 변수의 시각화</span>

```python
plt.figure(figsize=(10,10))
sns.barplot(y='중상자수', data=df);
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='사고유형_대범주', y='중상자수', data=df);
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='사고유형_대범주', y='중상자수', hue='가해운전자남성여부', data=df);
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.barplot(x='가해운전자차종', y='음주사고여부', data=df);
```

## <span style="color:#004f9f"> 3. 분포 확인 그래프</span>

### <span style="color:#004f9f"> 3.1. distplot </span>

```python
sns.displot(x='피해운전자연령', kind='hist', rug=True, bins=50, height=7, data=df);
```

```python
sns.displot(x='피해운전자연령', kind='kde', rug=True, height=7, data=df);
```

```python
sns.displot(x='피해운전자연령', kind='ecdf', height=7, data=df);
```

```python
sns.displot(x='피해운전자연령', hue='가해운전자남성여부', kind='kde', rug=True, height=7, data=df);
```

### <span style="color:#004f9f"> 3.2. boxplot </span>

```python
plt.figure(figsize=(10,10))
sns.boxplot(y='가해운전자연령', data=df);
#q1 하위 25%
#q3 상위 25%
#점찍히는건 이상치
#위스커값이 최소,최대로 걸림, 최소최대값
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.boxplot(x='가해운전자차종', y='가해운전자연령', data=df);
```

```python
col = ['승용','이륜','화물']
g_data = df[df['가해운전자차종'].isin(col)]

plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.boxplot(x='가해운전자차종', y='가해운전자연령', hue='피해운전자상해정도', data=g_data)
plt.legend(loc=2, bbox_to_anchor=(1,1));
```

### <span style="color:#004f9f"> 3.3. violinplot </span>

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.violinplot(x='가해운전자차종', y='가해운전자연령', data=g_data);
# 점이 중앙값
# 진한 검정선이 iqr
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.violinplot(x='가해운전자차종', y='가해운전자연령', hue='음주사고여부', split=True, data=g_data);
```

## <span style="color:#004f9f"> 4. 다변수의 관계 확인</span>

### <span style="color:#004f9f"> 4.1. jointplot</span>

```python
g_data = df[df['중상자수']>2]
sns.jointplot(x='중상자수', y='피해운전자연령', kind='scatter', height=10, data=g_data);
```

```python
sns.jointplot(x='중상자수', y='피해운전자연령', kind='hist', height=10, data=g_data);
```

```python
sns.jointplot(x='중상자수', y='피해운전자연령', kind='reg', height=10, data=g_data);
```

```python
sns.jointplot(x='중상자수', y='피해운전자연령', kind='kde', shade=True, height=10, data=g_data);
```

```python
sns.jointplot(x='중상자수', y='피해운전자연령', kind='hex', height=10, data=g_data);
```

### <span style="color:#004f9f"> 4.2. lmplot</span>

```python
col = ['승용','이륜','화물']
g_data = g_data[g_data['피해운전자차종'].isin(col)]
sns.lmplot(x='중상자수', y='피해운전자연령', hue='피해운전자차종', height=10, data=g_data);
```
