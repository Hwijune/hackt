```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='malgun gothic')
plt.rc('axes', unicode_minus=False)
%matplotlib inline
```

```python
weather = pd.read_csv('data/weather.csv', encoding='euc-kr')
weather.head()
```

```python
weather['월'] = pd.to_datetime(weather['월'], format='%b').dt.month
```

```python
weather_pivot = weather.pivot(index='월', columns='도시', values='평균기온')
weather_pivot
```

```python
weather_pivot.plot()
plt.title('도시별 평균기온')
plt.xlabel('월')
plt.ylabel('기온');
```

```python
weather_pivot.plot(subplots=True, figsize=(6,6));
```
