```python
import pymysql
import pandas as pd
```

```python
#Data Loading
artist_info = pd.read_csv('artist_info.csv')
```

```python
#DB INSERT 함수
def insert_row(cursor, data, table):
    col = ', '.join(data.index)
    place_holders = ', '.join(['%s']*len(data.index))
    key_holders = ', '.join([k+'=%s' for k in data.index]) 
    que = 'INSERT INTO {} ({}) VALUES ({}) ON DUPLICATE KEY UPDATE {}'.format(table, col, place_holders, key_holders)
    cursor.execute(que, list(data.values)*2)
```

```python
#DB 접속
host = 'host'
port = 3306
username = 'username'
database = 'production'
pw = 'pw'

conn = pymysql.connect(host=host, user=username, passwd=pw,
                      db=database, port=port,
                      use_unicode=True, charset='utf8')
cursor = conn.cursor()
```

```python
# 데이터 입력
for i, row in artist_info.iterrows():
    insert_row(cursor, row, 'Artists')
conn.commit()
```

```python
# 데이터 입력 확인
cursor.execute('SELECT * FROM Artists')
cursor.fetchall()
```

```python
# 데이터프레임으로 SQL 결과 가져오기
df = pd.read_sql('SELECT * FROM Artists', conn)
df.head()
```

```python
# DB접속 종료
cursor.close()
```
